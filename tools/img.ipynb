```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('bench2/build/gcc-o2/result.csv', index_col=0)
```

```python
algorithms = data.index
data_ranges = data.values
print(algorithms)
print(data_ranges)
```

```python
# 创建 x 轴数据
x = range(1, 18 + 1)

plt.figure(figsize=(8, 6))
plt.grid(True)
plt.rcParams["font.family"] = "sans"
# plt.rcParams["font.family"] = "Libertinus Sans"

# 绘制折线图
for i in range(len(algorithms)):
    plt.plot(x, data_ranges[i], marker='.', label=algorithms[i])

# 添加标题和标签
plt.title('Algorithm Performance')
plt.xlabel('Length')
plt.ylabel('Performance/ns')
plt.xticks(range(0, 19 + 1, 2))

# 添加图例
plt.legend()

# 显示图形
# plt.show()
plt.savefig('1.png', dpi=210)
```
